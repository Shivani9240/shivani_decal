```python
#Introduction to Matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def make_sine_wave(x, A, w):
    return A*np.sin(w*x)

x = np.linspace(0, 2*np.pi, 1000)
amplitudes = np.array([0.5, 1, 1.5, 2, 2.5])
frequencies = np.array([1, 2, 3, 4, 5])

plt.figure(figsize = (14,6))

for A, w in zip(amplitudes, frequencies):
    y = make_sine_wave(x, A, w)
    plt.plot(x, y, label=f"A={A}, w={w}")

plt.title("Sine Waves with Different Amplitudes and Frequencies")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
```

```python
#Data with Pandas
df = pd.read_csv("stars.csv")

#Print
print(df.head())
print(df.shape)
print(df.dtypes)

#Calculate
avg_mass = df["Mass (M☉)"].mean()
avg_temp = df["Temperature (K)"].mean()
print("Average mass: " + str(avg_mass) + " solar masses")
print("Average temperature: " + str(avg_temp) + " Kelvin")

largest_radius = df.loc[df["Radius (R☉)"].idxmax()]
print("Star with the largest radius: " + str(largest_radius))

M_type = df[df["Spectral_Type"].str.startswith("M")]
number_M_type = len(M_type)
print("Number of M-type stars: " + str(M_type))

#Sorting Stars
closest_stars = df.sort_values("Distance (ly)").head(3)
print("Three closest stars" + str(closest_stars))

M_type.to_csv("m_type_stars.csv", index = False)
```

```python
#Seaborn!
penguins = sns.load_dataset("penguins")
fig, axes = plt.subplots(1, 2, figsize = (10,5))

#Left Plot
sns.scatterplot(
    data = penguins, 
    x = "body_mass_g", 
    y = "bill_length_mm", 
    hue = "species", 
    ax = axes[0])
axes[0].set_title("Bill Length vs. Body Mass For Different Penguin Species")
axes[0].set_xlabel("Body Mass (g)")
axes[0].set_ylabel("Bill Length (mm)")
axes[0].legend(title="Species")
axes[0].grid(True)

#Right Plot
sns.histplot(
    data = penguins,
    x = "body_mass_g",
    bins=10,
    ax=axes[1])
axes[1].set_title("Distribution of Penguin Body Mass")
axes[1].set_xlabel("Body Mass (g)")
axes[1].set_ylabel("Number")
axes[1].grid(True)
```
